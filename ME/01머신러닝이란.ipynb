### 머신러닝 이란?
   y = 2x    [ 1,2,3,4,5]
: 수 많은 데이터를 학습(반복 계산)시켜 특정한 패턴(함수식)을 찾아내고 
  이를 기반으로 데이터를 분류(Classification)하거나 예측(Prediction).

- 일반소프트웨어(결정론적 소프트웨어)는 언제나 똑같은 과정을 거쳐 똑같은
  결과를 냅니다

- 머신러닝 기반 소프트웨어(비결정론적 소프트웨어)는 데이터와 알고리즘에 따라
  때로는 다른 결과를 추론해 냅니다. 
  (추론 : 이미 알고 있는 것으로부터 논리적 결론을 도출하는 과정)


### 작업 방법
- 어느 정도 규모가 있는 샘플 데이터를 입력으로 넣어 분석 시킨다
- 데이터에서 일정한 일정한 규칙을 찾아낸다(학습)
  이러면 AI하나 만든거다 AI 별거아니다
- 찾아낸 규칙을 기반드로 다른 데이터를 분류하거나 예측하는 것

### 어떻게 특징과 규칙을 찾을까?
- 특정 추출 데이터가 어떤 특징을 가지고 있는지 찾고 벡터로 만드는 처리과정
   어떤 특징(Feature)을 추출할지는 분석가가 결정. 이것이 머신러닝의 중요 포인트
- 회귀분석(y = ax + b) 데이터가 연속된 값일때 이를 선형회귀라고 부른다
  이때 Y를 종속변수(Target), X를 독립변수(Feature)라고 한다


### 머신러닝의 종류
### 지도학습(Supervised Learining)
: 학습시 데이터와 답을 같이 학습한 후 다른 데이터의 답을 예측하는 방법.
### 지도학습(Unsupervised Learining)
: 학습시 에이터만 학습한 후 다른 데이터의 규칙성을 찾습니다.
  - 클러스터 분석(Cluster Analysis)> 카데고리라고 생각하면 됨
  - 주성분 분석(Principal Component Analysis)
### 강화학습(Reinforcement Learning)
- 부분적으로 답을 입력한 후 데이터른 기반으로 최적의 답을 찾아 낸다( 로봇)

### 머신러닝의 흐름
(1)데이터 수집
(2)데이터 가공 및 분석
(3)데이터 학습
(3-1) 학습 방법 선택
(3-2) 매개변수(Hyper parameter) 조정 > 반복문 사용 , 차트로 확인
(3-3) 모델 학습
(4) 모델 평가
(5) 예측력과 오분류표가 좋은 경유 다음 단계 실행하고 아니면 (3)으로 재실행
(6) 성공

### 머신러닝 응용 분야
#### 클래스 분류(Classification)
- 특정 데이터에 레이블(Target)을 붙여 분류 할 수 있다.
- 스팸 메일 분류, 필기 인식, 증권 사기
##### 이진분류
- 둘 중 하나의 값으로 분류하는 경우 
##### 다중 분류
- 여러개의 분류값 중에서 하나의 값으로 예측하는 경우   ( 확율값 나온다)

#### 클러스터링(Clustering) : 그룹 나누기
- 값의 유사성을 기반으로 데이터를 여러 그룹으로 나눌 수 있다.
- 사용자의 취향을 그룹으로 묶어 사용자 취향에 맞는 광고를 제공.

#### 추천(Recommendation)
- 특징 데이터를 기반으로 다른 데이터를 추천하는 것.
- 인터넷 서점에서 구매한 책들을 기반으로 다른 책을 추천.

#### 차원 축소(Dimensionality Reduction)
- 데이터의 특성을 유지하면서 데이터 양을 줄이는 것
- 데이터를 시각화하거나 구조를 추출해서 용량을 줄여 빠르게 하거나 메모리 절약

### 프로젝트 과정
#### 문제 정의
- 프로젝트의 목표를 정의하고 그에 따른 계획을 세우는 단계
- 대략적으로 어떤 데이터를 수집해서 어떤 머신러닝 알고리즘으로 문제를
  해결할지 결정하는 단계
#### 데이터 획득
- 머신러닝 모델 학습에 필요한 데이터를 수집하는 단계
- 성곡적인머신러닝 프로젝트는 뛰어난 머신러닝 실력 보다도 얼마나 충분한
  데이터를 획득했는냐에 의해 결정
- 획득하 데이터는 학습(Training)데이터, 검증(valid)데이터, 평가(Test)데이터로 구분하여 사용

### 모델 구현
- 학습 데이터를 기반으로 한개 이상의 머신러닝 모델을 구현

### 검증
- 검증 데이터를 사용해 구현된 머신러닝 모델들의 성능을 검증
- 최종적으로 검증결과가 가장 높은 모델 하나를 선택

### 평가
- 검증 결과를 통해 선택된 모델의 일반화 선능을 확인.
- 평가 결과에 따라 프로젝트를 마무리할지 아니면 이전 단계로 돌아가서 개선할지를 결정

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# 데이터 로드
data = load_iris()
X = data.data

# PCA 적용 (2차원으로 축소)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
# print(X_pca)

# 시각화
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data.target)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA of Iris Dataset")
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# 1. 데이터 로드
data = load_iris()
X = data.data
y = data.target

# 2. PCA로 2차원 축소
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# 3. 시각화
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')

# 4. 범례 표시
plt.legend(handles=scatter.legend_elements()[0], labels=data.target_names)
plt.grid(True)
plt.show()
```
