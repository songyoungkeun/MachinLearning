```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split( iris_dataset['data'], iris_dataset['target'], stratify=iris_dataset['target'], random_state= 42)
```

```python
import pandas as pd
iris_datafrme = pd.DataFrame(train_input, columns= iris_dataset['feature_names'])
    
```

```python
import matplotlib.pyplot as plt


pd.plotting.scatter_matrix(
    iris_datafrme,
    c= train_target,
    figsize=(13,13),
    marker='o', # ^, *, +
    s = 100,
    alpha=0.7
)

plt.show()
```

```python
# KNN으로 예측하기
# 내주변의 1개만 선택하기
# test_input으로 에측력  구하기
# SL : 5, SW:2.9, PL:1 ,PW:0.2 일경우
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier(n_neighbors=5)

```

```python
kn.fit(train_input,train_target)
```

```python
kn.score(test_input,test_target)
```

```python
kn.score(train_input,train_target)
```

```python
# k값 찾기
```

```python
for i in range(1, 112 , 2) :
    kn = KNeighborsClassifier(n_neighbors=i)
    kn.fit(train_input, train_target)
    score = kn.score(test_input, test_target)
    print(F"{i}:{score}")
```

```python
def knnF(i):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(train_input, train_target)
    train_score = knn.score(train_input, train_target)
    test_score = knn.score(test_input, test_target)
    print(f"{i} : {train_score}, {test_score},\t{ train_score - test_score}")
```

```python
for i in range(1, 15,2):
        knnF(i)
```
