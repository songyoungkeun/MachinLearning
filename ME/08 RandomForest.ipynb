```python
import pandas as pd
```

```python
iris = pd.read_csv("../Data/iris.csv")
iris.head()
```

```python
# Feature와 Target 분리
iris_data = iris.loc[ : , 'SepalLength':'PetalWidth']  # 범위를 지정해서 쓸때는 대괄호쓰면 안됨  ['SepalLength':'PetalWidth'] > 'SepalLength':'PetalWidth'
iris_data.head()
```

```python
iris_target = iris.loc[ : , 'Name']
iris_target.head()
```

```python
# 확인또 확인하고 넘어가라  > 전체 갯수 확인
print(iris_data.shape)
print(iris_target.shape)
```

```python
a = 1
aa = (1,)  # 튜플

print(type(a))
print(type(aa)) #tuple은 컬렉션으로 read only list 이다
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_target, random_state= 42, stratify=iris_target)  # stratify 는 분류에서만 쓴다

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape )
```

```python
# sampling 판단 (sampling은 뽑기가아니고  모집단의 성격을 가져와야한다)
# descibe
print(X_train.describe())
print(X_test.describe())
```

```python
# RF로 분류
from sklearn.ensemble import RandomForestClassifier
```

```python
# 모델 만들기
clf = RandomForestClassifier(n_estimators=1) # n_estimator = 100 < 기본값
```

```python
# 힉습 시키기
clf.fit(X_train, y_train)
```

```python
# 평가하기   >  위에 올라가 n_estimator 숫자를 조절하면서 두개의 스코어가 좋은것 고른다
print(clf.score(X_train , y_train))
print(clf.score(X_test  , y_test))
```

```python
import pandas as pd
mr = pd.read_csv("../Data/mushroom.csv",header=None)
mr.head()
```

```python
# 한중리 버섯한종률
# 첫번재 열 : p(독버섯)>poison, e(식용) >eatable
# 두번째 열 : 벗섯의 머리 모양 => b(벨형태), c(원뿔모양), x(복록), f(평평), k(혹모양), s(오목)
# 세번쩨 열 : 버섯의 미리 색깔 => n(갈색), b(황갈색), c(연한갈색), ....
# 일단 각각의 기호가알파벳1자리 -> 숫자로 변환
# 연습 : 알파벳 => ASCII
ord('x')
# ASCII => 알파벳
chr(120)
# 연습 : mr의 1번 컬럼을 숫자로 변경
```

```python
target =[]
data = []

for row_index, row in mr.iterrows():   # 레코드 단위로 불러옴
    # print(row_index,":",row)
    # print(row.loc[0], ":", row.loc[1])
    target.append(row.loc[0])
    row_data = []
    for v in row.loc[1:]:
        row_data.append(ord(v))
    data.append(row_data)                                   
```

```python
# DataFrame 만들기
# Target을 기준으로 DataFrame 만들기
targetTemp = pd.DataFrame(target)
targetTemp.head()
```

```python
# DataFrame 만들기
# Target을 기준으로 DataFrame 만들고  Column Name을 겹치지 않게 하기
dataTemp = pd.DataFrame(data)
dataTemp.rename(
    columns= lambda x: x+1  ,   # 익명함수는 리턴값이 없다
    inplace=True
)
dataTemp.head()
```

```python
# 2개의 dataframe 합치기
mr2 = pd.concat([targetTemp,dataTemp], axis='columns')
mr2.head()
```
